```python
num_cols = []
cat_cols = []

for column in data.columns:
    if pd.api.types.is_numeric_dtype(data[column]):
        num_cols.append(column)
    else:
        cat_cols.append(column)
```

#### Univariate Analysis

```python
def numerical_univariate_analysis(column):
    fig, ax = plt.subplots(1,3,figsize=(10,5))
    sns.histplot(data[column], kde=True, bins=20, color='black', ax=ax[0])
    ax[0].set_title(f'Histogram of {column}')

    sns.boxplot(x=data[column], ax=ax[1])
    ax[1].set_title(f'Boxplot of {column}')

    sns.violinplot(x=data[column], ax=ax[2])
    ax[2].set_title(f'Violinplot of {column}')

    plt.show()
```

```python
for column in num_cols:
    print(f'Univariate Analysis for {column}')
    numerical_univariate_analysis(column)
```

```python
def categorical_univariate_analysis(column):
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))

    sns.countplot(data=data,x=data[column], palette='viridis', ax=ax[0])
    ax[0].set_title(f'Countplot for {column}')

    data_counts = data[column].value_counts()
    ax[1].pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
    ax[1].set_title(f'Pie plot for {column}')
    
    plt.show()
```

```python
for col in cat_cols:
    print(f' Univariate analysis for {col}')
    categorical_univariate_analysis(col)
```

```python
def bivariate_num_col(col):
    sns.set_palette("tab10")
    
    fig, ax = plt.subplots(1, 2, figsize=(20, 10))

    sns.regplot(x=data[col], y=data['price'], ax=ax[0])
    ax[0].set_title(f'Regression Plot: {col} vs Price')

    sns.scatterplot(x=col, y='price', data=data, ax=ax[1])
    ax[1].set_title(f'Scatter Plot of {col} vs price')
    
    plt.tight_layout()
    plt.show()
```

```python
for col in num_cols[1:]:
    print(f'Bivariate analysis between {col} and price')
    bivariate_num_col(col)
```

```python
def bivariate_cat(col):
    fig, ax = plt.subplots(1, 2, figsize=(20, 10))
    sns.barplot(x=col, y='price', data=data, ax=ax[0])
    ax[0].set_title(f'Average Price by {col}')

    sns.boxplot(x=col, y='price', data=data, ax=ax[1])
    ax[1].set_title(f'Price Distribution by {col}')
    plt.tight_layout()
    plt.show()
```

```python
for col in cat_cols[1:]:
    print(f'Bivariate analysis between {col} and price')
    bivariate_cat(col)
```
